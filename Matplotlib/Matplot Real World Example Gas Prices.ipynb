# Importing Required Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Importing data

```python
gas = pd.read_csv(r"C:\Users\HP\Desktop\AI and ML\gas_prices.csv")

print(gas)
```

## Plotting Starts !!

```python
plt.plot(gas.Year , gas.USA , label="USA" )
plt.plot(gas['Year'] , gas.Canada , label="Canada" )
plt.plot(gas.Year , gas['South Korea'] , label="South Korea")
plt.legend()
plt.show()
```

### Editing above graph and making it presentable

```python
plt.figure(figsize=(10,5))


plt.title('Gas Prices in different countries')


plt.xlabel('Year')
plt.ylabel('$/Gallon')

plt.plot(gas.Year , gas.USA,'b.-', label='USA' )
plt.plot(gas['Year'] , gas.Canada,'g.-', label='Canada' )
plt.plot(gas.Year , gas['South Korea'] , 'r.-',label='South Korea')


plt.xticks(gas.Year[::3])

plt.legend()
plt.show()
```

### Plotting all countries

```python
plt.figure(figsize=(11,7))
for country in gas:
    if(country!='Year'):
        plt.plot(gas.Year,gas[country],marker='.', label=country)
        
plt.xticks(gas.Year[::3])

plt.title('Gas Prices in different countries')


plt.xlabel('Year')
plt.ylabel('$/Gallon')


plt.legend()

plt.savefig('Gas_Prices_Graph.png', dpi=300)
plt.show()
        
```

# Working on Fifa Data

## Loading Data

```python
fifa=pd.read_csv(r"C:\Users\HP\Desktop\AI and ML\fifa_data.csv")
fifa.head(5)
```

### Histograms

```python
plt.hist(fifa.Overall)
plt.show()
```

### Editing the graph

```python
bins=[40,50,60,70,80,90]

plt.title('Overall Skills of fifa players', fontdict={'fontweight':'bold','fontsize':20})

plt.hist(fifa.Overall,bins=bins,color='orange')


plt.xticks(bins)

plt.xlabel('Overall Skills(%)')
plt.ylabel('Number of players')
plt.savefig("Overall Skill Histogram.png",dpi=300)

plt.show()
```

 ### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

plt.title('Preferred Foot of FIFA players')

labels=['Left Foot', 'Right Foot']
color=['wheat','bisque']

plt.pie([left,right],labels=labels,colors=color , autopct='%.2f %%')

plt.savefig("Foot Reference.png",dpi=300)

plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 160].count()[0]
medium = fifa.loc[(fifa.Weight < 170) & (fifa.Weight>= 160)].count()[0]
heavy = fifa.loc[fifa.Weight > 170].count()[0]

plt.title("Weight Distribution of FIFA Players(in LBS)")

labels=['Light Weight(<160)','Medium Weight(160-170)', 'Heavy Weight(>170)']
plt.pie([light,medium,heavy],labels=labels,autopct="%.2f %%")

plt.savefig("Weight Distribution.png",dpi=300)

plt.show()
```

## Box Plot(shows comparison between two or more datasets)

```python
barcelona = fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club=='Real Madrid']['Overall']
revs = fifa.loc[fifa.Club=='New England Revolution']['Overall']
chelsea = fifa.loc[fifa.Club == 'Chelsea']['Overall']


plt.figure(figsize=(6,10))


plt.title("Professional team comparison")

plt.ylabel("FIFA Overall Rating")


labels=['FC Barcelona','Real Madrid','England Revolution','Chelsea']
plt.boxplot([barcelona,madrid,revs,chelsea],labels=labels)

plt.savefig("overall Comparison.png",dpi=300)
plt.show()
```
