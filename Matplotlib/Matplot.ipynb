```python
import matplotlib.pyplot as plt
import numpy as np

x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y)
plt.title("First Graph")
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.show()
```

# Changing X and Y axis' numbers and adding legend

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y,label='2x', marker="*",markersize=10,markeredgecolor='green',linestyle="--")
plt.title("First Graph")
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

##----Adding Legend----##:
plt.legend()
plt.show()
```

### Shorthand notation

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

# Format =[color][marker][line type]
plt.plot(x,y,'g^--',label='2x')


plt.title("First Graph")
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()
```

### Plotting 2 graphs on same axes

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

# Format =[color][marker][line type]
plt.plot(x,y,'g^--',label='2x')

'''----------------------'''
x2=np.arange(0,4,0.5)
print(x2)
plt.plot(x2,x2**2,'y',label='x^2')
'''----------------------'''
plt.title("First Graph")
plt.xlabel('X - axis')
plt.ylabel('Y - axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()
```

### Plotting Subparts , resizing and saving the graph


```python
x2=np.arange(0,4,0.5)

# format = plt.figure(figsize(x:y ratio), dpi={preffered value>300})  dpi= density of pixels per inch
plt.figure(figsize=(5,3),dpi=300)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')

plt.legend()

#----------------------------------
plt.savefig('mygraph.png', dpi=400)
#_---------------------------------
plt.show()
```
